### Variação percentual no número de professores e alunos (2019 - 2022)

Fazer o download das tabelas "2.10 - Evolução do quadro de Docentes por regime de trabalho" e "2.03 - Evolução do número de Alunos matriculados de Graduação, por Unidade" no site https://uspdigital.usp.br/anuario/AnuarioControle# e substituir o caminho na célula abaixo.

```python
#Variação no número de Docentes (2019 - 2022)

docentes = pd.read_excel(r"C:\Users\Diogo\Downloads\T2.10.xls")
docentes = docentes.dropna()
docentes = docentes.loc[(docentes['Evolução do quadro de Docentes por regime de trabalho'] == 'A - Ensino e Pesquisa')]
dezenove = docentes.groupby('Unnamed: 2').get_group(2019)
vinte = docentes.groupby('Unnamed: 2').get_group(2020)
vinte_um = docentes.groupby('Unnamed: 2').get_group(2021)
vinte_2 = docentes.groupby('Unnamed: 2').get_group(2022)

dezenove = dezenove.groupby('Unnamed: 1').sum()
vinte = vinte.groupby('Unnamed: 1').sum()
vinte_um = vinte_um.groupby('Unnamed: 1').sum()
vinte_2 = vinte_2.groupby('Unnamed: 1').sum()

queda = ((vinte_2['Unnamed: 4'] - dezenove['Unnamed: 4'])/dezenove['Unnamed: 4'])
queda = pd.DataFrame(queda)

#Variação no número de alunos (2019 - 2022)

alunos = pd.read_excel(r"C:\Users\Diogo\Downloads\T2.03.xls")
alunos = alunos.dropna()
alunos = alunos.loc[(alunos['Evolução do número de Alunos matriculados de Graduação, por Unidade'] == 'A - Ensino e Pesquisa')]
alunos_2022 = alunos.groupby('Unnamed: 2', as_index = True).get_group(2022)
alunos_2019 = alunos.groupby('Unnamed: 2', as_index = True).get_group(2019)
alunos_2022 = alunos_2022.drop(['Unnamed: 2', 'Evolução do número de Alunos matriculados de Graduação, por Unidade'], axis = 1)
alunos_2019 = alunos_2019.drop(['Unnamed: 2', 'Evolução do número de Alunos matriculados de Graduação, por Unidade'], axis = 1)
alunos_2022 = alunos_2022.set_index('Unnamed: 1')
alunos_2019 = alunos_2019.set_index('Unnamed: 1')

#Tabela com variações

queda_alunos = ((alunos_2022['Unnamed: 3'] - alunos_2019['Unnamed: 3'])/alunos_2019['Unnamed: 3'])
queda['Alunos'] = queda_alunos
```

```python
#Formatação

queda.loc[:, "Alunos"] = queda["Alunos"].map('{:.2%}'.format)
queda.loc[:, "Unnamed: 4"] = queda["Unnamed: 4"].map('{:.2%}'.format)
queda['Alunos'] = queda['Alunos'].str.replace('.', ',')
queda['Unnamed: 4'] = queda['Unnamed: 4'].str.replace('.', ',')
display(queda)
```

**Salvar dados em .xlsx**

```python
queda.to_excel('queda no numero de profs e alunos.xlsx')
```

## Proporções étnicas

Fazer o download da tabela "2.21 - Alunos de graduação, pós-graduação e pós-doutorandos distribuídos por Unidade e raça/cor" no site https://uspdigital.usp.br/anuario/AnuarioControle# e substituir o caminho na célula abaixo.

```python
raça = pd.read_excel(r"C:\Users\Diogo\Downloads\T2.21.xls")
raça = raça.dropna()
raça = raça.loc[(raça['Unnamed: 2'] == 'Graduação')]
raça = raça.loc[(raça['Alunos de graduação, pós-graduação e pós-doutorandos distribuídos por Unidade e raça/cor'] == 'A - Ensino e Pesquisa')]
alunos = raça.groupby('Unnamed: 1')
alunos = alunos.sum()
```

```python
#Encontrar proporções

etnia = pd.DataFrame()
amarelo = raça.loc[(raça['Unnamed: 3'] == 'Amar')]
amarelo.set_index('Unnamed: 1', inplace = True)
brancos = raça.loc[(raça['Unnamed: 3'] == 'Bran')]
brancos.set_index('Unnamed: 1', inplace = True)
indigenas = raça.loc[(raça['Unnamed: 3'] == 'Indig')]
indigenas.set_index('Unnamed: 1', inplace = True)
pardos = raça.loc[(raça['Unnamed: 3'] == 'Pard')]
pardos.set_index('Unnamed: 1', inplace = True)
pretos = raça.loc[(raça['Unnamed: 3'] == 'Pret')]
pretos.set_index('Unnamed: 1', inplace = True)
ni = raça.loc[(raça['Unnamed: 3'] == 'N/I')]
ni.set_index('Unnamed: 1', inplace = True)

prop = amarelo['Unnamed: 4']/alunos['Unnamed: 4']
etnia = pd.concat([etnia, prop], axis = 1)
etnia = etnia.rename(columns={"Unnamed: 4": 'Amarelo'})
    
prop = brancos['Unnamed: 4']/alunos['Unnamed: 4']
etnia = pd.concat([etnia, prop], axis = 1)
etnia = etnia.rename(columns={"Unnamed: 4": 'Brancos'})
    
prop = indigenas['Unnamed: 4']/alunos['Unnamed: 4']
etnia = pd.concat([etnia, prop], axis = 1)
etnia = etnia.rename(columns={"Unnamed: 4": 'Indígenas'})
    
prop = pardos['Unnamed: 4']/alunos['Unnamed: 4']
etnia = pd.concat([etnia, prop], axis = 1)
etnia = etnia.rename(columns={"Unnamed: 4": 'Pardos'})
    
prop = pretos['Unnamed: 4']/alunos['Unnamed: 4']
etnia = pd.concat([etnia, prop], axis = 1)
etnia = etnia.rename(columns={"Unnamed: 4": 'Pretos'})
    
prop = ni['Unnamed: 4']/alunos['Unnamed: 4']
etnia = pd.concat([etnia, prop], axis = 1)
etnia = etnia.rename(columns={"Unnamed: 4": 'N/I'})
```

```python
total_pardos = pardos['Unnamed: 4'].sum()
total_pretos = pretos['Unnamed: 4'].sum()
total_ind = indigenas['Unnamed: 4'].sum()
total = indigenas['Unnamed: 4'].sum() + pretos['Unnamed: 4'].sum() + pardos['Unnamed: 4'].sum() + brancos['Unnamed: 4'].sum() + amarelo['Unnamed: 4'].sum() + ni['Unnamed: 4'].sum()

pre = total_pretos/total
par = total_pardos/total
ind = total_ind/total
prop = (pre + par + ind)*100

print(f'A proporção de PPIs na USP é de {prop:.2f}%')
```

```python
#Formatação

etnia.loc[:, "Amarelo"] = etnia["Amarelo"].map('{:.2%}'.format)
etnia.loc[:, "Brancos"] = etnia["Brancos"].map('{:.2%}'.format)
etnia.loc[:, "Indígenas"] = etnia["Indígenas"].map('{:.2%}'.format)
etnia.loc[:, "Pardos"] = etnia["Pardos"].map('{:.2%}'.format)
etnia.loc[:, "Pretos"] = etnia["Pretos"].map('{:.2%}'.format)
etnia.loc[:, "N/I"] = etnia["N/I"].map('{:.2%}'.format)
etnia['Amarelo'] = etnia['Amarelo'].str.replace('.', ',')
etnia['Indígenas'] = etnia['Indígenas'].str.replace('.', ',')
etnia['Pardos'] = etnia['Pardos'].str.replace('.', ',')
etnia['Pretos'] = etnia['Pretos'].str.replace('.', ',')
etnia['Brancos'] = etnia['Brancos'].str.replace('.', ',')
etnia['N/I'] = etnia['N/I'].str.replace('.', ',')
```

**Salvar dados em .xlsx**

```python
etnia.to_excel('etnia.xlsx')
```

**Unidades com mais proporção de determinada etnia**

Basta substituir a etnia na célula abaixo

```python
etnia.sort_values(by=['Pardos'])
```

## Gastos com terceirizados

Fazer o download da busca da dotação orçamentária detalhada por Unidade no portal https://uspdigital.usp.br/portaltransparencia/financeiro

```python
#Gasto com terceirizados

total_gasto = pd.read_csv(r"C:\Users\Diogo\Downloads\Orcamento_2023 (3).csv", sep=';', encoding='latin-1', thousands = '.', decimal = ',')
terceirizados = total_gasto.loc[(total_gasto['Elemento'] == '339039 - Outros Serviços de Terceiros - PJ') | (total_gasto['Elemento'] == '339037 - Serviços Limpeza, Vigilância e Outros - PJ')]
terceirizados = terceirizados.groupby('Unidade de Despesa').sum()

#Gasto total

total_gasto = total_gasto.groupby('Unidade de Despesa').sum()
terceirizados['Dotação total'] = total_gasto['Dotação Atual (R$)']

#Percentual com terceirizados

terceirizados['Percentual'] = (terceirizados['Dotação Atual (R$)']/terceirizados['Dotação total'])
```

```python
#Formatação

terceirizados.loc[:, "Dotação Atual (R$)"] = 'R$ ' + terceirizados["Dotação Atual (R$)"].map('{:.2f}'.format)
terceirizados.loc[:, "Dotação total"] = 'R$ ' + terceirizados["Dotação total"].map('{:.2f}'.format)
terceirizados.loc[:, "Percentual"] = terceirizados["Percentual"].map('{:.2%}'.format)
terceirizados['Dotação Atual (R$)'] = terceirizados['Dotação Atual (R$)'].str.replace('.', ',')
terceirizados['Dotação total'] = terceirizados['Dotação total'].str.replace('.', ',')
terceirizados['Percentual'] = terceirizados['Percentual'].str.replace('.', ',')
display(terceirizados)
```

**Salvar dados em .xlsx**

```python
terceirizados.to_excel('gastos com terceirizados.xlsx', header  = ['Gastos com terceirizados', 'Gasto total', 'Percentual com terceirizados'])
```
